```python
import numpy as np
import math
import random as rand

class hold:
    #creats a hold. the left, right and top of the hold have a rating on how good it is to grab
    
    def __init__(self):
        self.left = 0
        self.right = 0
        self.top = 0

    def setLeft(self, left):
        self.left = left

    def setRight(self, right):
        self.right = right

    def setTop (self, top):
        self.top = top 

    def getHolds(self):
        print((self.left, self.top, self.right), end = '')
        
    def getSide(self, num):
        if num==0: return self.left
        elif num==1: return self.top
        elif num==2: return self.right
        else: return None
```

```python
#The distance and angle functions used for calculating the grade score
def angle(x_curr, x_next):
    #INPUT: x_curr and x_next are arrays containing the x,y coordinates of your current and next hold repectively
    #OUTPUT: array of length 2 containing the side that will be grabbed ([1,2,3] -> [left,right,top]) and the distance 
    x_dist = abs(x_curr[0] - x_next[0])
    y_dist = x_next[1] - x_curr[1] #x_next will always be equal or greater than x_curr (always moving up)
    dist = np.sqrt(x_dist**2 + y_dist**2) 
    if x_dist != 0 and np.arctan(y_dist/x_dist) >= np.pi/4:
        #grabbing from the top
        return [1,dist]
    else:
        if x_curr[0] <= x_next[0]:
            #grabbing the right side
            return [2,dist]
        #grabbing the left side
        return [0,dist]

#the weight function to be ran with the angle output and the hold score
def Weight(score, hold):
    #input: score: angle function output, hold: the hold object of the hold being moved too
    #output: a score as a float
    #Need to add:
    #    taking into account where the other hand is
    #    matching? (or maybe that gets put into the other part of the code)
    return hold.getSide(score[0]) + score[1]
```
